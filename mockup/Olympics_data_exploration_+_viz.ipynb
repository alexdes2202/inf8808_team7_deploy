# Age par Categorie

```python
df = olympics_data[olympics_data["Sport"] == "Swimming"].copy()

bins = [10, 14, 17, 20, 23, 26, 30, 35, 100]
labels = ["10-14", "15-17", "18-20", "21-23", "24-26", "27-30", "31-35", "36+"]
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)


df_grouped = df.groupby(["Year", "Age Group", "Event"]).size().reset_index(name="Count")

color_palette = ["#2E86C1", "#1ABC9C", "#F39C12", "#E74C3C", "#8E44AD", "#16A085", "#D35400", "#34495E"]

app = dash.Dash(__name__)

app.layout = html.Div([
    html.Label("Select Event:", style={"font-weight": "bold", "font-size": "16px"}),
    dcc.Dropdown(
        id="event-dropdown",
        options=[{"label": event, "value": event} for event in df_grouped["Event"].unique()],
        value=df_grouped["Event"].unique()[0],
        clearable=False,
        style={"width": "50%", "margin-bottom": "20px"}
    ),

    # Bubble Chart
    dcc.Graph(id="bubble-chart")
])


@app.callback(
    Output("bubble-chart", "figure"),
    [Input("event-dropdown", "value")]
)
def update_chart(selected_event):
    filtered_df = df_grouped[df_grouped["Event"] == selected_event]
    fig = px.scatter(filtered_df,
                     x="Year",
                     y="Age Group",
                     size="Count",
                     color="Age Group",
                     color_discrete_sequence=color_palette,
                     title=f"Age Distribution of {selected_event} Over Time",
                     labels={"Year": "Year", "Age Group": "Age Group", "Count": "Number of Athletes"},
                     opacity=0.85,
                     size_max=40)
    fig.update_layout(
        template="simple_white",
        font=dict(size=14),
        title_font=dict(size=18),
        plot_bgcolor="white",
        paper_bgcolor="white",
        xaxis=dict(showline=False, showgrid=False, tickmode="array"),
        yaxis=dict(showline=False, showgrid=True, gridcolor="lightgray", gridwidth=1, griddash="dashdot"),
        legend_title_text="Age Groups"
    )

    # Reverse the y-axis categories (Age Group)
    fig.update_layout(
        yaxis=dict(
            categoryorder='category descending'  # Reverse the order of the y-axis categories
        )
    )

    return fig

# Run app
if __name__ == "__main__":
    app.run_server(debug=True)
```

# Participations

```python
# A améliorer

df = olympics_data[olympics_data["Sport"] == "Swimming"].copy()
df = df.drop_duplicates(subset=["Name", "Year"])

df = df.sort_values(["Name", "Year"])
df["Participation_Number"] = df.groupby("Name").cumcount() + 1

df["Medal_Status"] = df["Medal"].apply(lambda x: "Medal Won" if pd.notna(x) else "No Medal")

participation_counts = df.groupby(["Participation_Number", "Medal_Status"]).size().unstack(fill_value=0)

participation_percentages = participation_counts.div(participation_counts.sum(axis=1), axis=0) * 100

colors = {"Medal Won": "#F4C542", "No Medal": "#A0A0A0"}

participation_levels = sorted(participation_percentages.index)
num_participations = len(participation_levels)

fig = sp.make_subplots(
    rows=1, cols=num_participations,
    specs=[[{"type": "domain"}] * num_participations]
)

for i, p in enumerate(participation_levels):
    values = [
        participation_percentages.loc[p]["Medal Won"],
        participation_percentages.loc[p]["No Medal"]
    ]

    fig.add_trace(go.Pie(
        labels=["Medal Won", "No Medal"],
        values=values,
        marker=dict(colors=[colors["Medal Won"], colors["No Medal"]]),
        hole=0.4,
        textinfo='percent',
        showlegend=(i == 0)
    ), row=1, col=i+1)

    num_pies = len(participation_levels)
    x_position = (i + 0.5) / num_pies

    fig.add_annotation(
        x=x_position,
        y=0.5,
        text=f"{p}",
        showarrow=False,
        font=dict(size=16, color="black", family="Arial Black"),
        xref="paper",
        yref="paper"
    )

fig.update_layout(
    title_text="Medal Distribution by Olympic Participation",
    height=400,
    width=130 * num_participations,
    margin=dict(t=80, b=50, l=50, r=50),
    font=dict(size=12)
)

fig.show()
```

```python
import dash
from dash import dcc, html
import plotly.graph_objects as go
import pandas as pd
from dash.dependencies import Input, Output

# Assuming olympics_data is your dataset
# Get the unique years available in the data
available_years = olympics_data['Year'].unique()

# Function to create the Sankey plot for a given year or for all editions
def create_sankey_plot(year):
    # If the selected year is "All Editions", we include all years
    if year == "All Editions":
        df_medals = olympics_data[(olympics_data["Sport"] == "Swimming")]
    else:
        df_medals = olympics_data[(olympics_data["Sport"] == "Swimming") & (olympics_data["Year"] == year)]

    # Compter les participations par pays, uniquement pour les médailles (sans NaN)
    df_medals_with_medals = df_medals[df_medals['Medal'].notna()]
    total_medal_counts = df_medals_with_medals['NOC'].value_counts()

    # Sélectionner le Canada + les 3 meilleurs pays
    top_countries = total_medal_counts.head(3).index.tolist()
    selected_country = 'CAN'  # Canada
    if selected_country not in top_countries:
        top_countries.append(selected_country)

    # Filtrer les données pour ne garder que les pays sélectionnés
    df_medals = df_medals[df_medals['NOC'].isin(top_countries)]

    # Créer une nouvelle colonne pour regrouper les médailles
    df_medals['Medal'] = df_medals['Medal'].fillna('No Medal')

    # Créer une colonne Medal_NOC pour chaque pays et type de médaille
    df_medals['Medal_NOC'] = df_medals['Medal'] + '_' + df_medals['NOC']
    df_medals['No_Medal_NOC'] = 'No Medal_' + df_medals['NOC']  # No Medal spécifique pour chaque pays

    # Compter le nombre de médailles par pays et médaille
    medal_counts = df_medals.groupby(['NOC', 'Medal', 'Medal_NOC', 'No_Medal_NOC']).size().reset_index(name='Count')

    # Trier les pays par le nombre total de médailles décroissant
    total_counts_sorted = medal_counts.groupby('NOC')['Count'].sum().sort_values(ascending=False)
    sorted_countries = total_counts_sorted.index.tolist()

    # Réordonner les données selon les pays triés
    medal_counts['NOC'] = pd.Categorical(medal_counts['NOC'], categories=sorted_countries, ordered=True)
    medal_counts = medal_counts.sort_values('NOC')

    # Liste des nœuds (nodes) pour le Sankey
    countries = medal_counts['NOC'].unique().tolist()

    # Ajouter les nœuds pour "Medal" et "No Medal" spécifiques pour chaque pays
    all_labels = countries + [f'{medal}_{country}' for country in countries for medal in ['Gold', 'Silver', 'Bronze']] + [f'No Medal_{country}' for country in countries]

    # Mapping des indices pour les pays et les deux catégories (Medal et No Medal)
    source_indices = []
    target_indices = []
    values = []

    for country in countries:
        # Compter le nombre total de médailles et de "No Medal" pour chaque pays
        # medal_count = medal_counts[medal_counts['NOC'] == country].shape[0]
        medal_count = df_medals[(df_medals['NOC'] == country) & (df_medals['Medal'] != 'No Medal')].shape[0]
        no_medal_count = len(df_medals[df_medals['NOC'] == country]) - medal_count

        # Ajouter les liens entre le pays et les catégories de médaille spécifiques
        for medal in ['Gold', 'Silver', 'Bronze']:
            medal_NOC = f'{medal}_{country}'
            count = medal_counts[(medal_counts['Medal_NOC'] == medal_NOC)]['Count'].sum()
            if count > 0:  # Si le pays a gagné cette médaille
                source_indices.append(all_labels.index(country))
                target_indices.append(all_labels.index(medal_NOC))
                values.append(count)

        # Ajouter les liens pour les pays qui n'ont pas de médaille
        if no_medal_count > 0:
            no_medal_NOC = f'No Medal_{country}'
            source_indices.append(all_labels.index(country))
            target_indices.append(all_labels.index(no_medal_NOC))
            values.append(no_medal_count)

    # Gestion des couleurs : rouge pour Canada, bleu pour les autres pays, gris pour "No Medal"
    node_colors = []

    # Couleurs pour les médailles
    medal_colors = {
        'Gold': '#FFD700',  # Or
        'Silver': '#C0C0C0',  # Argent
        'Bronze': '#CD7F32',  # Bronze
        'No Medal': '#FFFFFF'  # Gris pour No Medal
    }

    # Ajouter des couleurs pour les pays
    for label in countries:
        if label == selected_country:
            node_colors.append('red')  # Rouge pour le Canada
        else:
            node_colors.append('black')  # Bleu pour les autres pays

    # Ajouter des couleurs pour les médailles et "No Medal"
    for label in all_labels:
        if 'Gold' in label:
            node_colors.append(medal_colors['Gold'])
        elif 'Silver' in label:
            node_colors.append(medal_colors['Silver'])
        elif 'Bronze' in label:
            node_colors.append(medal_colors['Bronze'])
        elif 'No Medal' in label:
            node_colors.append(medal_colors['No Medal'])


    # Création du diagramme de Sankey
    fig = go.Figure(go.Sankey(
        textfont=dict(color="rgba(0,0,0,0)", size=1),
        orientation = 'v',
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color='black', width=0.5),
            label=all_labels,
            color=node_colors

        ),
        link=dict(
            source=source_indices,
            target=target_indices,
            value=values
        )
    ))

    # Mise en forme
    fig.update_layout(
        title_text=f'Medal distribution in Swimming : ({year}) (Count)',
        font_size=12
    )

    return fig


def create_sankey_plot_percent(year):
    if year == "All Editions":
        df_medals = olympics_data[(olympics_data["Sport"] == "Swimming")]
    else:
        df_medals = olympics_data[(olympics_data["Sport"] == "Swimming") & (olympics_data["Year"] == year)]

    df_medals_with_medals = df_medals[df_medals['Medal'].notna()]
    total_medal_counts = df_medals_with_medals['NOC'].value_counts()

    top_countries = total_medal_counts.head(3).index.tolist()
    selected_country = 'CAN'
    if selected_country not in top_countries:
        top_countries.append(selected_country)

    df_medals = df_medals[df_medals['NOC'].isin(top_countries)]
    df_medals['Medal'] = df_medals['Medal'].fillna('No Medal')

    df_medals['Medal_NOC'] = df_medals['Medal'] + '_' + df_medals['NOC']
    df_medals['No_Medal_NOC'] = 'No Medal_' + df_medals['NOC']

    medal_counts = df_medals.groupby(['NOC', 'Medal', 'Medal_NOC', 'No_Medal_NOC']).size().reset_index(name='Count')

    total_counts_per_country = df_medals.groupby('NOC').size()  # Total participations per country

    medal_counts['Percentage'] = medal_counts.apply(lambda row: (row['Count'] / total_counts_per_country[row['NOC']]) * 100, axis=1)  # Normalize to percentage

    total_counts_sorted = medal_counts.groupby('NOC')['Count'].sum().sort_values(ascending=False)
    sorted_countries = total_counts_sorted.index.tolist()

    medal_counts['NOC'] = pd.Categorical(medal_counts['NOC'], categories=sorted_countries, ordered=True)
    medal_counts = medal_counts.sort_values('NOC')

    countries = medal_counts['NOC'].unique().tolist()
    all_labels = countries + [f'{medal}_{country}' for country in countries for medal in ['Gold', 'Silver', 'Bronze']] + [f'No Medal_{country}' for country in countries]

    source_indices = []
    target_indices = []
    values = []

    for country in countries:
        for medal in ['Gold', 'Silver', 'Bronze']:
            medal_NOC = f'{medal}_{country}'
            percentage = medal_counts[(medal_counts['Medal_NOC'] == medal_NOC)]['Percentage'].sum()
            if percentage > 0:
                source_indices.append(all_labels.index(country))
                target_indices.append(all_labels.index(medal_NOC))
                values.append(percentage)  # Store percentage instead of count

        no_medal_count = 100 - sum(values[-3:])  # Ensure total is 100%
        if no_medal_count > 0:
            no_medal_NOC = f'No Medal_{country}'
            source_indices.append(all_labels.index(country))
            target_indices.append(all_labels.index(no_medal_NOC))
            values.append(no_medal_count)

    node_colors = []
    medal_colors = {'Gold': '#FFD700', 'Silver': '#C0C0C0', 'Bronze': '#CD7F32', 'No Medal': '#FFFFFF'}

    for label in countries:
        node_colors.append('red' if label == selected_country else 'black')

    for label in all_labels:
        if 'Gold' in label:
            node_colors.append(medal_colors['Gold'])
        elif 'Silver' in label:
            node_colors.append(medal_colors['Silver'])
        elif 'Bronze' in label:
            node_colors.append(medal_colors['Bronze'])
        elif 'No Medal' in label:
            node_colors.append(medal_colors['No Medal'])

    fig = go.Figure(go.Sankey(
        orientation='v',
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color='black', width=0.5),
            label=all_labels,
            color=node_colors
        ),
        link=dict(
            source=source_indices,
            target=target_indices,
            value=values  # Now contains percentages
        )
    ))

    fig.update_layout(
        title_text=f'Medal distribution in Swimming : ({year}) (%)',
        font_size=12
    )

    return fig


# Initialize Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    # Dropdown to select the year
    html.H1("Olympic Swimming Medal Distribution"),

    # Dropdown to select relative or absolute values
    dcc.Dropdown(
        id='scale-dropdown',
        options=[
            {'label': 'Relative', 'value': 'percent'},
            {'label': 'Absolute', 'value': 'absolute'}
        ],
        value='percent',  # Default to relative (percent)
        style={'width': '50%'}
    ),

    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in available_years] + [{'label': 'All Editions', 'value': 'All Editions'}],
        value='All Editions',  # Default to 'All Editions'
        style={'width': '50%'}
    ),

    # Graph to display the Sankey plot
    dcc.Graph(id='sankey-plot')
])


# Callback to update the Sankey plot based on selected year
@app.callback(
    Output('sankey-plot', 'figure'),
    [Input('year-dropdown', 'value'),
     Input('scale-dropdown', 'value')]
)
def update_sankey_plot(selected_year, scale):
    if scale == 'percent':
        return create_sankey_plot_percent(selected_year)
    else:
        return create_sankey_plot(selected_year)


# Run the Dash app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import plotly.graph_objs as go
import pandas as pd


df = swimming_part_counts_detailed[swimming_part_counts_detailed['Participation_Number'] <= 4]


df_melted = df.melt(id_vars=['Sport', 'Participation_Number'],
                    value_vars=['Gold_Percentage', 'Silver_Percentage', 'Bronze_Percentage'],
                    var_name='Percentage_Label',
                    value_name='Percentage_Value')

df_melted['label'] = df_melted['Participation_Number'].astype(str) + df_melted['Percentage_Label']
df_melted['Medal_Label'] = df_melted['Percentage_Label'].apply(lambda x: x.replace('_Percentage', ''))


df_melted = df_melted.sort_values(by=['Participation_Number', 'Percentage_Value'], ascending=[True, False])



df_melted['x_position'] = df_melted.groupby('Percentage_Label').cumcount() + 1  # Rank 1, 2, 3


"""
traces = [
    go.Bar(
        y=df['Gold_Percentage'], x=df['Participation_Number'], name='Gold',
        marker=dict(color='gold')
    ),
    go.Bar(
        y=df['Silver_Percentage'], x=df['Participation_Number'], name='Silver',
        marker=dict(color='silver')
    ),
    go.Bar(
        y=df['Bronze_Percentage'], x=df['Participation_Number'], name='Bronze',
        marker=dict(color='#cd7f32')  # Bronze color
    ),
]
"""

traces = []


for index, row in df.iterrows():
    
    percentages = {
        'Gold': row['Gold_Percentage'],
        'Silver': row['Silver_Percentage'],
        'Bronze': row['Bronze_Percentage']
    }

    
    sorted_percentages = sorted(percentages.items(), key=lambda x: x[1], reverse=True)

    
    first, second, third = sorted_percentages

    
    x_positions = {
        first[0]: 1,    
        second[0]: 0,   
        third[0]: 2     
    }

    participation_traces = [
        go.Bar(
            y=[percentages[medal]],
            x=[row['Participation_Number'] + x_positions[medal] * 0.2],  
            name=medal,
            marker=dict(color={'Gold': 'gold', 'Silver': 'silver', 'Bronze': '#cd7f32'}[medal]),
            showlegend=True if index == 0 else False 
        ) for medal, _ in sorted_percentages
    ]
    traces.extend(participation_traces)

    medal_emojis = {
        'Gold': '🏅',    
        'Silver': '🥈',  
        'Bronze': '🥉'    
    }

    for medal, pos in zip(['Gold', 'Silver', 'Bronze'], [first, second, third]):
        traces.append(
            go.Scatter(
                x=[row['Participation_Number'] + x_positions[medal] * 0.2],
                y=[percentages[medal] + 0.2],  
                text=[medal_emojis[medal]],  
                mode="text",
                showlegend=False 
            )
        )

# Create the layout for the chart
"""
layout = go.Layout(
    barmode='group',
    title='Podium of Gold, Silver, and Bronze Percentages per Participation',
    yaxis=dict(title='Percentage (%)'),
    xaxis=dict(title='Participation Number'),
    height=600
)
"""

layout = go.Layout(
    title="Podium of Gold, Silver, and Bronze Winning Chances per Participation (Swimming)",
    xaxis=dict(
        title="Participation Number",
        tickvals=df['Participation_Number'],  
        ticktext=df['Participation_Number'].astype(str)  
    ),
    yaxis=dict(title="Percentage (%)"),
    barmode="group", 
)

fig = go.Figure(layout=layout)

"""
color_map = {
    'Gold': 'gold',
    'Silver': 'silver',
    'Bronze': '#cd7f32'
}
"""
color_map = ['gold', 'silver', '#cd7f32']

fig.add_traces(
    traces
)

fig.show()
```
